```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


import pandas as pd
import os


for i in range(1982, 2024 ):
    
    url=f"https://statiz.sporki.com/stats/?m=main&m2=fielding&m3=default&so=&ob=&year={i}&sy=&ey=&te=&po=&lt=10100&reg=A&pe=&ds=&de=&we=&hr=&ha=&ct=&st=&vp=&bo=&pt=&pp=&ii=&vc=&um=&oo=&rr=&sc=&bc=&ba=&li=&as=&ae=&pl=&gc=&lr=&pr=1000&ph=&hs=&us=&na=&ls=1&sf1=G&sk1=&sv1=&sf2=G&sk2=&sv2="

    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    temp = soup.find_all("table")[0]
    temp2 = temp.find_all("tr")

    column=[]

    for tr in temp2:
    
        if tr.find("th"):
            continue
        else :
            row = []
            tds = tr.find_all("td")
            for c in tds:
                row.append(c.get_text())
            print(row)
            column.append(row)
            
    df= pd.DataFrame(data=column,columns=["순위","이름","team","WAAwithPOS","출장", "선발", "이닝","수비기회","자살","보살","실책","수비율","RF9","rangeRAA","Arm RAA","Err RAA","DP RAA","Bt RAA","CS RAA","Blk RAA", "Frm RAA","종합 RAA","종합 RAA /144","POSAdj", "RAAwithPOS","WAAwoPOS","WAAwithPOS"])
    df.to_csv(f"baseball_stats_{i}.csv", index=False, encoding='utf-8-sig')



# 모든 CSV 파일을 저장할 빈 DataFrame 생성
combined_df = pd.DataFrame()

# 파일 경로 설정
folder_path = "."

# 폴더 내 모든 파일에 대해 반복
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # CSV 파일을 DataFrame으로 읽기
        df = pd.read_csv(os.path.join(folder_path, filename))
        # 모든 DataFrame을 결합
        combined_df = pd.concat([combined_df, df], ignore_index=True)

# 하나의 CSV 파일로 결합된 DataFrame을 CSV로 내보내기
combined_df.to_csv("combined_baseball_stats.csv", index=False, encoding='utf-8-sig')

```
